```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy.io import loadmat
```

```python
path = '../CRWU'

mat_files = glob.glob(os.path.join(path, '**', '*.mat'), recursive=True)
```

```python
import os
import scipy.io

# Structure optimisée pour stocker les données
df = {
    "File_Path": [],
    "RPM_Level": [],
    "Fault_Type": [],
    "Condition": [],
    "Severity": [],
    "File_Name": [],
    "Data": [],
    "Shape": [],
}

# Récupérer tous les fichiers .mat dans le dossier CRWU
mat_files = [
    os.path.join(root, file)
    for root, _, files in os.walk('../CRWU')
    for file in files if file.endswith('.mat')
]

# Traitement des fichiers
for file in mat_files:
    
    try:
        data = scipy.io.loadmat(file)
        
        parts = file.split(os.sep)
        print(parts)
        rpm_level = parts[1].split(" ")[0] if len(parts) > 1 else None
        fault_type = parts[2] if len(parts) > 3 else None
        condition = parts[3] if fault_type == 'Outer Race' else None
        severity = parts[-2] if len(parts) > 3 else None
        file_name = os.path.basename(file)

        # Parcourir les clés utiles du fichier
        for key, value in data.items():
            if key.startswith('__'):
                continue  
            
            if hasattr(value, 'shape') and value.shape[0] > 1:
                value = np.squeeze(value)
                df["Data"].append(value.tolist() if isinstance(value, (list, tuple)) else value)
                df["Shape"].append(value.shape[0])
                df["File_Path"].append(file)
                df["RPM_Level"].append(rpm_level)
                df["Fault_Type"].append(fault_type)
                df["Condition"].append(condition)
                df["Severity"].append(severity)
                df["File_Name"].append(file_name)
        
    except Exception as e:
        print(f"Erreur lors du traitement du fichier {file} : {e}")
```

```python
for v in df.values():
    print(len(v))
```
